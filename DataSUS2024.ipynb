# Task
Clean the provided dataset.

## Handle missing values

### Subtask:
Address the missing values in the columns 'Nome Instituição', 'Inserção', 'Unidade Fornecimento', and 'ANVISA'. This could involve dropping rows or columns, or imputing missing values based on the nature of the data in each column.


**Reasoning**:
Calculate and display the percentage of missing values for the specified columns to inform the decision on how to handle them.



```python
missing_percentage = dados[['Nome Instituição', 'Inserção', 'Unidade Fornecimento', 'ANVISA']].isnull().mean() * 100
display(missing_percentage)
```

**Reasoning**:
Based on the missing value percentages, 'ANVISA' has a high percentage of missing values (over 60%), making imputation or dropping rows less effective. 'Nome Instituição' and 'Unidade Fornecimento' have very low percentages, so dropping rows with missing values in these columns is a reasonable approach to retain most of the data. 'Inserção' has a moderate percentage (around 14%), and since its date-related, imputing with the median is a suitable strategy to preserve the distribution.



```python
# Drop the 'ANVISA' column due to a high percentage of missing values
dados = dados.drop('ANVISA', axis=1)

# Drop rows with missing values in 'Nome Instituição' and 'Unidade Fornecimento'
dados = dados.dropna(subset=['Nome Instituição', 'Unidade Fornecimento'])

# Impute missing values in 'Inserção' with the median
if pd.api.types.is_numeric_dtype(dados['Inserção']):
    median_insercao = dados['Inserção'].median()
    dados['Inserção'].fillna(median_insercao, inplace=True)
else:
    # If 'Inserção' is not numeric, impute with mode or a placeholder
    # Check data type to determine appropriate imputation
    mode_insercao = dados['Inserção'].mode()[0] if not dados['Inserção'].mode().empty else 'Unknown'
    dados['Inserção'].fillna(mode_insercao, inplace=True)

# Verify that missing values have been addressed in the target columns
print("Missing values after cleaning:")
display(dados[['Nome Instituição', 'Inserção', 'Unidade Fornecimento']].isnull().sum())
```

## Check data types

### Subtask:
Verify that the data types of each column are appropriate for analysis and convert them if necessary (e.g., convert 'Preço Unitário' and 'Preço Total' to numeric types).


**Reasoning**:
Display the data types of all columns to identify columns that need type conversion.



```python
display(dados.dtypes)
```

**Reasoning**:
Convert 'Preço Unitário' and 'Preço Total' columns to numeric by removing commas and converting to float, coercing errors to NaN. Then display the data types again to confirm the conversion.



```python
# Convert 'Preço Unitário' to numeric
dados['Preço Unitário'] = dados['Preço Unitário'].astype(str).str.replace(',', '.', regex=False)
dados['Preço Unitário'] = pd.to_numeric(dados['Preço Unitário'], errors='coerce')

# Convert 'Preço Total' to numeric
dados['Preço Total'] = dados['Preço Total'].astype(str).str.replace(',', '.', regex=False)
dados['Preço Total'] = pd.to_numeric(dados['Preço Total'], errors='coerce')

# Display data types to confirm conversion
display(dados.dtypes)
```

### Datetime conversion

The code cell below converts the 'Compra' and 'Inserção' columns to datetime objects.

- `pd.to_datetime()` is used for the conversion.
- `origin='1899-12-30'` is specified because the original numeric values appear to be OLE Automation dates, which use this date as their origin.
- `unit='D'` indicates that the numeric values represent the number of days since the origin.
- `errors='coerce'` is used for the 'Inserção' column to turn any values that cannot be parsed into datetime objects into `NaT` (Not a Time).

```python
print("Data types:")
display(dados[['Compra', 'Inserção']].dtypes)

print("\nSample values from 'Compra':")
display(dados['Compra'].head())

print("\nSample values from 'Inserção':")
display(dados['Inserção'].head())
```

```python
# Convert 'Compra' and 'Inserção' to datetime objects
dados['Compra'] = pd.to_datetime(dados['Compra'], origin='1899-12-30', unit='D')
dados['Inserção'] = pd.to_datetime(dados['Inserção'], origin='1899-12-30', unit='D', errors='coerce')

print("Data types after conversion:")
display(dados[['Compra', 'Inserção']].dtypes)

print("\nSample values from 'Compra' after conversion:")
display(dados['Compra'].head())

print("\nSample values from 'Inserção' after conversion:")
display(dados['Inserção'].head())
```
